# Logistic Regression Plots

## Logistic Function


```python
import numpy as np
import matplotlib.pyplot as plt
from myst_nb import glue
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
Z = np.linspace(-5, 5, 50)
sZ = sigmoid(Z)

plt.rcParams.update(plt.rcParamsDefault)

plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
fig = plt.figure(figsize=plt.figaspect(0.33))
ax = fig.add_subplot(111)

# Horizontal grid line at 0.5
major_ticks = np.arange(0, 101, 0.5)
ax.set_yticks(major_ticks)
ax.yaxis.grid(True, ls='--', lw=0.5)

# Vertical grid line at z = 0
major_ticks = np.arange(-5,5.1,1)
ax.set_xticks(major_ticks)
ax.xaxis.grid(True, ls='--', lw=0.5)

ax.plot(Z, sZ, c='r')

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

glue("sigmoid_example", fig, display=True)


```
